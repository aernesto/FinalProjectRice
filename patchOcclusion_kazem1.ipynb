```python
#from sklearn.feature_extraction import image
#neighbors = image.extract_patches_2d(img, (3, 3))
import numpy as np
a = np.arange(80).reshape((8,10))
print(a)
a_padded = np.lib.pad(a, [(2,2),(2,2)], 'constant', constant_values=(0, 0))
print(a_padded)
w = 2
h = 2

print(a_padded.nonzero())
print(np.min(a_padded.nonzero(), axis = 1))
print(np.max(a_padded.nonzero(), axis = 1))
print(a_padded[2,2])
print(a_padded[2,6])
print(a_padded[6,2])
print(a_padded[6,6])

max_r = np.max(a_padded.nonzero(), axis = 1)[0]
max_c = np.max(a_padded.nonzero(), axis = 1)[1]
min_r = np.min(a_padded.nonzero(), axis = 1)[0]
min_c = np.min(a_padded.nonzero(), axis = 1)[1]

strides1 = 2
strides2 = 2
for i in range(min_r,max_r-w+2, strides1):
    for j in range(min_c, max_c-h+2, strides2):
        a_tmp = a_padded.copy()
        a_tmp[i:i+w, j:j+h] = 0
        print("index: ", i, j)
        print(a_tmp)
        print(2*'\n')
        
```
